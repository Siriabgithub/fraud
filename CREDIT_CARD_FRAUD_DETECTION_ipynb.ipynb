BUILDING A BALANCED SAMPLE DATASET INVOLVING AN UNDER-SAMPLING TECHNIQUE, I SELECTED A SUBSET OF THE ORIGINAL DATASET THAT ENSURES A SIMILAR DISTRIBUTION OF NORMAL TRANSACTIONS AND FRAUDULENT TRANSACTIONS. SPECIFICALLY, THE SAMPLE DATASET COMPRISES 492 FRAUDULENT TRANSACTIONS, AIMING TO ADDRESS IMBALANCES IN THE ORIGINAL DATA FOR MORE ACCURATE ANALYSIS AND MODEL TRAINING.








```python
legit_sample = legit.sample(n=492)
```

NOW WE WILL
CONCATENATING TWO DATAFRAMES THAR WE CREATE BEFORE

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

AGAIN WE SEE TEH FIRST 5 DATA OF THE DATA SET THE NEW DATA SET

```python
new_dataset.head()
```

AGAIN WE SEE TEH LAST 5 DATA OF THE DATA SET THE NEW DATA SET

```python
new_dataset.tail()
```

AGAIN DISTRIBUTION OF LEGIT TRANSACTIONS & FRAUDULENT TRANSACTIONS

```python
new_dataset['Class'].value_counts()
```

AGAIN TO COMPARE THE VALUES FOR BOTH TRANSACTIONS WE WILL WRITE THIS CODE (new_dataset.groupby('Class').mean())

```python
new_dataset.groupby('Class').mean()
```

ASA A NEXT STEP WE WILL SPLITTING THE DATA INTO FEATURES & TARGETS

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

NOW THE DATA SET IS REDDY TO SPLITTING  INTO TRAINING DATA & TESTING DATA

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

AFTER SPLITING THE DATA SET WE CAN THE SHAPE BY USING print(X.shape, X_train.shape, X_test.shape).

```python
print(X.shape, X_train.shape, X_test.shape)
```


LOGISTIC REGRESSION IS HIGHLY SUITABLE FOR THIS PROJECT DUE TO ITS SIMPLICITY, EFFICIENCY, AND INTERPRETABILITY. IT EXCELS IN BINARY CLASSIFICATION PROBLEMS, EFFECTIVELY DISCERNING BETWEEN LEGITIMATE AND FRAUDULENT TRANSACTIONS. FURTHERMORE, LOGISTIC REGRESSION OFFERS PROBABILITIES, FACILITATING RISK ASSESSMENT FOR EFFECTIVE FRAUD IDENTIFICATION.

```python
model = LogisticRegression()
```

AFTER SELECTION THE LOGISTIC REGRESSION MODEL TFANNING THE MODEL WITH OUR TRAINING DATA

```python
model.fit(X_train, Y_train)
```

AFTER TRANNING THE DATA WE WILL CHECK THE ACCURACY ON OUR TRAINING DATA

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

AFTER TRANNING THE DATA WE WILL CHECK THE ACCURACY ON OUR TEST DATA

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
